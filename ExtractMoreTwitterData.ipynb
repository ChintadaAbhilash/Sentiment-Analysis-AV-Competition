```
from twitterscraper import query_tweets
import datetime as dt
import pandas as pd
```

**Get Politics Data**

```
begin_date = dt.date(2017 , 9 , 1)
end_date = dt.date(2019 , 9 , 2)

limit = 10000

lang = 'english'

tweets = query_tweets("#politics" , begindate = begin_date , enddate = end_date , limit = limit , lang = lang)
```

```
Df = pd.DataFrame(t.__dict__ for t in tweets)
```

```
Df.head()
```

```
pol_final = Df[["username" , "text"]]
```

```
pol_final.head()
```

```
pol_final.shape
```

**Tech News**

```
begin_date = dt.date(2017 , 9 , 1)
end_date = dt.date(2019 , 9 , 2)

limit = 10000

lang = 'english'

tweets = query_tweets("#technology" , begindate = begin_date , enddate = end_date , limit = limit , lang = lang)
```

**Sports**

**Concatenate all DataFrames**

```
Final = pd.concat([pol_final , tech_final , movie_final , sports_final] , axis = 0 , sort =False)
```
